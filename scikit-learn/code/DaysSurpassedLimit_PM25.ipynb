```python
import pandas as pd
data = pd.read_csv('PM25_2010-2014.csv')
```

```python
data
```

```python
data = data.bfill()
```

```python
for i , k in data.iterrows(): 
    print(k['Data fine'], k['PM2.5'])
```

```python
data_pm25 = pd.DataFrame({'data': data['Data fine'], 'pm25':data['PM2.5']})
data_pm25
```

```python
data_pm25['pm25'].mean()
```

```python
anno_i = []
for i, k in data_pm25.iterrows():
    anno = k['data'].split('/')[2]
    print(anno)
    anno_i.append(anno)
    
data_pm25['anno'] = anno_i
```

```python
data_pm25
```

```python
group_by_anno = data_pm25.groupby(data_pm25['anno'])
```

```python
for i, k in group_by_anno: 
    print(i, k)
```

```python
for i, k in group_by_anno: 
    counter = 1
    for j , n in k.iterrows():
        pm25_i = n['pm25']
       
        if pm25_i > 35:
            counter += 1
            print(str(n['data']) +", "+ str(n['pm25'])+", "+str(counter))
    counter = 1        
```

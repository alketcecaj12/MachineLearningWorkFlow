```python
import pandas as pd
```

```python
dati_pm10 = pd.read_csv("PM10_2010-2014.csv")
```

```python
dati_pm10
```

```python
dati_pm10['PM10'].mean()
```

```python
dati = pd.DataFrame({'data': dati_pm10['Data fine'], 'pm10': dati_pm10['PM10']})
dati
```

```python
dati = dati.bfill()
dati
```

```python
for i, k in dati.iterrows(): 
    pm10_gg = k['pm10']
    if pm10_gg > 50: 
        print(k['data'], k['pm10'])
    
```

```python
year = []
for i, k in dati.iterrows():
    dateformat = k['data'].split('/')
    print(dateformat[2])
    year.append(dateformat[2])                           
```

```python
dati['anno'] = year
dati
```

```python
for i, k in dati.iterrows(): 
    print(k['data'], k['anno'])
```

```python
group_by_anno = dati.groupby(dati['anno'])
```

```python
for i, k in group_by_anno: 
    #print(i, k)
    print(i, k['pm10'].mean())
```

```python
for i, k in group_by_anno: 
    dati_i = k
    counter = 1
    for j, row in dati_i.iterrows(): 
        pm10_gg = row['pm10']
        if pm10_gg > 50: 
            print(row['pm10'], row['anno'], counter)
            counter += 1
        
        
        
```
